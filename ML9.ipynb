```python
corpus = [
    ["I", "love", "dogs"],
    ["I", "love", "cats"],
    ["You", "love", "dogs"],
    ["You", "like", "cats"],
    ["We", "love", "dogs"]
]
```

```python
import numpy as np
from collections import defaultdict

# Build transition probabilities
transition_counts = defaultdict(lambda: defaultdict(int))

for sentence in corpus:
    for i in range(len(sentence) - 1):
        curr_word = sentence[i]
        next_word = sentence[i + 1]
        transition_counts[curr_word][next_word] += 1

# Convert to probabilities
transition_probs = {}

for curr_word, next_words in transition_counts.items():
    total = sum(next_words.values())
    transition_probs[curr_word] = {
        word: count / total for word, count in next_words.items()
    }

# Prediction function
def hmm_predict_next(word):
    if word not in transition_probs:
        return None
    return max(transition_probs[word], key=transition_probs[word].get)

# Test
print("HMM Prediction after 'love':", hmm_predict_next("love"))
print("HMM Prediction after 'You':", hmm_predict_next("You"))
```

```python
from collections import defaultdict

# Build PST
pst = defaultdict(lambda: defaultdict(int))

for sentence in corpus:
    for i in range(len(sentence) - 1):
        context = tuple(sentence[:i+1])  # growing context
        next_word = sentence[i+1]
        pst[context][next_word] += 1

# Convert to probabilities
pst_probs = {}

for context, next_words in pst.items():
    total = sum(next_words.values())
    pst_probs[context] = {
        word: count / total for word, count in next_words.items()
    }

# Prediction function
def pst_predict_next(context):
    context = tuple(context)

    if context not in pst_probs:
        return None

    return max(pst_probs[context], key=pst_probs[context].get)

# Test
print("PST Prediction after ['I','love']:", pst_predict_next(["I", "love"]))
print("PST Prediction after ['You']:", pst_predict_next(["You"]))
```
